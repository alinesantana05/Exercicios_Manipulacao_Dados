2️⃣ Filtre as vendas feitas apenas de "Notebook" e armazene em um novo DataFrame.

Crie uma nova coluna "Mês" extraída da coluna "Data_Venda" e agrupe os dados por mês, calculando o total de vendas em cada mês.

```python
import pandas as pd
```

```python
data = {'Produto' : ['Notebook','Mouse', 'Teclado', 'Monitor', 'Mesa', 'Câmera'],
        'Categoria' : ['Eletrônico', 'Acessório', 'Acessório', 'Acessório','Mobília', 'Acessório'],
        'Preço' : [3759, 50.9, 79.80, 382.7, 543, 42.3],
        'Quantidade' : [30, 55, 9, 65, 23, 90],
        'Data_Venda' : ['22-1-2022', '4-2-2022', '21-3-2022', '9-2-2022', '2-4-2022', '4-2-2022'], }
```

```python
data = pd.DataFrame(data)
data
```

```python
data.info()
```

```python
data['Data_Venda'] = pd.to_datetime(data['Data_Venda'])
data
```

```python
data['Total_Venda'] = (data['Preço'] * data['Quantidade'])
data
```

```python
filtro = data[data['Produto'] == 'Notebook']
filtro
```

```python
AgrColunas = data.groupby("Categoria")["Total_Venda"].sum()
AgrColunas
```

```python
data.loc[data['Quantidade'].idxmax(), ['Produto', 'Quantidade']]
```

```python
data.sort_values(by ='Total_Venda', ascending=False)
```

```python
data['Mês'] = data['Data_Venda'].dt.month
```

```python
resultado_Mês = data.groupby('Mês')['Total_Venda'].sum()
resultado_Mês
```
